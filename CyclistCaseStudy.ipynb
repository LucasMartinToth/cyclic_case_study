```python
#INSTALAR BIBLIOTECAS
import numpy as np
import pandas as pd
```

```python
#IMPORTAR DADOS DE PLANILHAS
q1_2019 = pd.read_csv("Divvy_Trips_2019_Q1.csv")
q1_2020 = pd.read_csv("Divvy_Trips_2020_Q1.csv")
```

```python
#ORGANIZAR DADOS E COMBINAR TABELAS

#Comparar colunas das duas tabelas
print("Colunas Q1 2019:\n", q1_2019.columns)
print("\nColunas Q1 2020:\n", q1_2020.columns)
```

```python
#Renomear colunas de 2019 para alinhas com 2020
q1_2019 = q1_2019.rename(columns={
'trip_id': 'ride_id',
'bikeid': 'rideable_type',
'start_time': 'started_at',
'end_time': 'ended_at',
'from_station_name': 'start_station_name',
'from_station_id': 'start_station_id',
'to_station_name': 'end_station_name',
'to_station_id': 'end_station_id',
'usertype': 'member_casual'
})
```

```python
#Inspecionar Dataframe 2019
print("\nQ1 2019 Data Info:")
q1_2019.info()
```

```python
#Inspecionar Dataframe 2020
print("\nQ1 2020 Data Info:")
q1_2020.info()
```

```python
#Converter ride_id e rideable_type de 2019 para Strings, para combinar com 2020 
q1_2019['ride_id'] = q1_2019['ride_id'].astype(str)
q1_2019['rideable_type'] = q1_2019['rideable_type'].astype(str)
```

```python
#Juntar os dois dataframes em um único
all_trips = pd.concat([q1_2019, q1_2020], ignore_index=True)

#Remover colunas lat, long, birthyear, and gender, pois não existem mais em 2020
all_trips = all_trips.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng', 'birthyear', 'gender',
'tripduration'], errors='ignore')
```

```python
#LIMPEZA DE DADOS PARA A ANÁLISE

#Inspecionar novo Dataframe
print("\nLista de nomes de colunas:\n", all_trips.columns)
print("\nNúmero de linhas no Dataframe:", len(all_trips))
print("\nDimensões do Dataframe:", all_trips.shape)
print("\n6 primeiras linhas do Dataframe:\n", all_trips.head())
print("\nVer colunas e tipos de dados:\n")
all_trips.info()
```

```python
#Verificar valores de member_casual
print("\nValores de 'member_casual' antes de limpeza:\n",
all_trips['member_casual'].value_counts())
```

```python
#Correção dos valores de member_casual
all_trips['member_casual'] = all_trips['member_casual'].replace({
'Subscriber': 'member',
'Customer': 'casual'
})
```

```python
#Verificar valores de member_casual
print("\nValores de 'member_casual' depois da limpeza:\n",
all_trips['member_casual'].value_counts())
```

```python
#Converter "started_at" e "ended_at" para formato Datetime
all_trips['started_at'] = pd.to_datetime(all_trips['started_at'])
all_trips['ended_at'] = pd.to_datetime(all_trips['ended_at'])
```

```python
#Adicionar colunas que listem a data, mês, dia e ano de cada viagem
all_trips['date'] = all_trips['started_at'].dt.date
all_trips['month'] = all_trips['started_at'].dt.month
all_trips['day'] = all_trips['started_at'].dt.day
all_trips['year'] = all_trips['started_at'].dt.year
all_trips['day_of_week'] = all_trips['started_at'].dt.day_name()

#Adicionar cálculo "ride_length" para todas as viagens (em minutos)
all_trips['ride_length'] = (all_trips['ended_at'] - all_trips['started_at']).dt.total_seconds() / 60
```

```python
#Inspecionar tabela com novos tipos de dados
print("\nTipos de dados depois de adicionar colunas de data e ride_length:\n")
all_trips.info()
```

```python
#Excluir viagens de manutenção, "viagens no tempo" e outliars, remover duplicatas
all_trips_v2 = all_trips[(all_trips['start_station_name'] != "HQ QR") & (all_trips['ride_length'] >=
0) & (all_trips['ride_length'] <= 1440)].copy()
all_trips_v2.drop_duplicates()
```

```python
#CONDUZIR ANÁLISE DESCRITIVA

#Duração da viagem (em minutos)
print("\nAnálise Descritiva da coluna ride_length:\n", all_trips_v2['ride_length'].describe())
```

```python
#Comparar usuários casuais com assinantes
print("\nDuração da viagem agrupada por member_casual:\n",
all_trips_v2.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min']))
```

```python
#Ver média de duração da viagem baseada no tipo de usuário e dia da semana
print("\nMédia de duração da viagem por tipo de usuário e dia da semana:\n",
all_trips_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].mean())
```

```python
#Ordenar dias da semana pra melhor análise
days_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
all_trips_v2['day_of_week'] = pd.Categorical(all_trips_v2['day_of_week'],
categories=days_order, ordered=True)
```

```python
#Ver média de duração da viagem baseada no tipo de usuário e dia da semana ordenado
print("\nMédia de duração da viagem por tipo de usuário e dia da semana:\n",
all_trips_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].mean())
```

```python
#Resumo das viagens e duração agrupadas por tipo de usuário e dia da semana
summary_stats = all_trips_v2.groupby(['member_casual', 'day_of_week']).agg(
number_of_rides=('ride_id', 'count'),
average_duration=('ride_length', 'mean')
).reset_index()
print("\nResumo das viagens e duração agrupadas por tipo de usuário e dia da semana:\n", summary_stats)
```

```python
#EXPORTAR .CSV PARA VISUALIZAÇÃO NO TABLEAU

counts = all_trips_v2.groupby(['member_casual', 'day_of_week']).agg(
number_of_rides=('ride_id', 'count'),
average_duration=('ride_length', 'mean')
).reset_index()
counts.to_csv('avg_duration.csv', index=False)
```
